```python
import pandas as pd
import numpy as np

# Chargement des vrais fichiers
df_transactions = pd.read_csv('../data/transactions.csv')
df_employes = pd.read_csv('../data/employes.csv')

# Vérification transactions
print("=== TRANSACTIONS ===")
print(f"Lignes     : {len(df_transactions)}")
print(f"Colonnes   : {list(df_transactions.columns)}")
print(f"Nulls      : {df_transactions.isnull().sum().sum()}")
print(f"Doublons   : {df_transactions.duplicated().sum()}")
print(df_transactions.head(3))

print("\n=== EMPLOYES ===")
print(f"Lignes     : {len(df_employes)}")
print(f"Colonnes   : {list(df_employes.columns)}")
print(f"Nulls      : {df_employes.isnull().sum().sum()}")
print(f"Doublons   : {df_employes.duplicated().sum()}")
print(df_employes.head(3))
```
